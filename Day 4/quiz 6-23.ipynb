```python
#20.แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่
group_0 = (df['Age'] >= 0) & (df['Age'] <= 10)
group_1 = (df['Age'] > 10 ) & (df['Age'] <= 20)
group_2 = (df['Age'] > 20 ) & (df['Age'] <= 30)
group_3 = (df['Age'] > 30 ) & (df['Age'] <= 40)
group_4 = (df['Age'] > 40 ) & (df['Age'] <= 50)
group_5 = (df['Age'] > 50 ) & (df['Age'] <= 60)
group_6 = (df['Age'] > 60 ) & (df['Age'] <= 70)
group_7 =(df['Age'] > 70 ) & (df['Age'] <= 80)
```

```python
group_0_all= df[group_0]['PassengerId'].count()
group_1_all= df[group_1]['PassengerId'].count()
group_2_all= df[group_2]['PassengerId'].count()
group_3_all= df[group_3]['PassengerId'].count()
group_4_all= df[group_4]['PassengerId'].count()
group_5_all= df[group_5]['PassengerId'].count()
group_6_all= df[group_6]['PassengerId'].count()
group_7_all= df[group_7]['PassengerId'].count()
```

```python
group_0_surv = df[(df['Survived']==1) & group_0 ]['PassengerId'].count()
group_1_surv = df[(df['Survived']==1) & group_1 ]['PassengerId'].count()
group_2_surv = df[(df['Survived']==1) & group_2 ]['PassengerId'].count()
group_3_surv = df[(df['Survived']==1) & group_3 ]['PassengerId'].count()
group_4_surv = df[(df['Survived']==1) & group_4 ]['PassengerId'].count()
group_5_surv = df[(df['Survived']==1) & group_5 ]['PassengerId'].count()
group_6_surv = df[(df['Survived']==1) & group_6 ]['PassengerId'].count()
group_7_surv = df[(df['Survived']==1) & group_7 ]['PassengerId'].count()
```

```python
print('กลุ่มอายุ 0-10 :{} %'.format (group_0_surv /group_0_all))
print('กลุ่มอายุ 10-20 : {} %'.format (group_1_surv /group_1_all))
print('กลุ่มอายุ 20-30 :{} %'.format (group_2_surv /group_2_all))
print('กลุ่มอายุ 30-40 :{} %'.format (group_3_surv /group_3_all))
print('กลุ่มอายุ 40-50 :{} %'.format( group_4_surv /group_4_all))
print('กลุ่มอายุ 50-60 :{} %'.format (group_5_surv /group_5_all))
print('กลุ่มอายุ 60-70 :{} %'.format (group_6_surv /group_6_all))
print('กลุ่มอายุ 70-80 :{} %'.format (group_7_surv /group_7_all))
```
