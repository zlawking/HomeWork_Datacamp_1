```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
df
```

```python
#9. ให้หาอัตราส่วนตั๋วโดยสารชั้น 1 2 3 และพล็อตเป็น Pie chart
df1=df['Pclass'].value_counts()
df1
```

```python
class1_ratio = df1[1] / df1.sum()
class2_ratio = df1[2] / df1.sum()
class3_ratio = df1[3] / df1.sum()
ratio = [class1_ratio,class2_ratio,class3_ratio]
```

```python
plt.figure(figsize = [5, 5])
labels = ['Pclass1', 'PClass2', 'PClass3']
plt.pie(ratio, labels = labels)
```

```python
#10.ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้Queenstown เป็น explode
df2=df['Embarked'].value_counts()
df2
```

```python
ratio2 = [df2['S'],df2['C'],df2['Q']]
```

```python
plt.figure(figsize = [5, 5])
labels = ['Southampton', 'Cherbourg', 'Queenstown']
explode = [0, 0, 0.1]
plt.pie(ratio2, labels = labels,explode = explode)
```

```python
#11. นำราคาตั๋วโดยสารทั้งหมดมาหา histogram
fare = df['Fare']
plt.hist(fare, bins=30)
```

```python
#12.สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y
age = df['Age']
fare = df['Fare']
plt.scatter(age, fare)
```

```python
#13.สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น
class1 = df[df['Pclass'] == 1]['Fare'].mean()
class2 = df[df['Pclass'] == 2]['Fare'].mean()
class3 = df[df['Pclass'] == 3]['Fare'].mean()
labels = ['PClass 1', 'PClass 2', "PClass 3"]
plt.bar(['PClass1', 'PClass2', 'PClass3'], [class1, class2, class3])
```

```python
#14. สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร
ratio_age = df['Fare'] / df['Age']
age = df['Age']
plt.bar(age, ratio_age)
```

```python
#15.สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y
df['last_name'] = df['Name'].str.split(',').str[0]
df1 = df.groupby('last_name').count()
```

```python
df2=df1.groupby('PassengerId').count()
number_same=df2['Survived']
number_same
```

```python
make_in = number_same[number_same.index > 1]
lst_in = list(make_in.index)
lst_num = list(make_in.values)
plt.bar(lst_in, lst_num)
```

```python
#16(Optional) สร้างหลายกราฟในรูปเดียว โดยมีสองคอลัมน์ คอลัมน์แรกมีสามแถว คอลัมน์สองมีสองแถว พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน โดยสมการแต่ละกราฟสามารถคิดเองได้โดยใช้ข้อมูลจากไฟล์ titanic และไม่ต้องซ้ำกัน
plt.figure(figsize=[6,6])
fig = plt.figure()
sup1 = fig.add_subplot(321, title="Popular price")
x = df['Fare']
y = df['PassengerId']
sup1.set_xlabel('Fare')
sup1.set_ylabel('Passenger')
sup1.scatter(x, y)

sup2 = fig.add_subplot(322, title="Popular age group")
x2 = df['Age']
y2 = df['PassengerId']
sup2.set_ylabel('Passenger')
sup2.set_xlabel('Age')
sup2.scatter(x2, y2)

sup3 = fig.add_subplot(323, title="Age group that survived")
x3 = df['Age']
y3 = df['Survived']
sup3.set_xlabel('Age')
sup3.set_ylabel('Survived')
sup3.scatter(x3, y3)


ax4 = fig.add_subplot(324, title="Number of people(Sex)")
x4 = df['Sex']
ax4.set_xlabel('Sex')
ax4.set_ylabel('Number of people ')
plt.hist(x4,bins=2)


ax5 = fig.add_subplot(325, title="Number of people(Class)")
x5 = df['Pclass']
ax5.set_xlabel('Pclass')
ax5.set_ylabel('Number of people ')
plt.hist(x5,bins=3)


fig.set_figheight(20)
fig.set_figwidth(20)
plt.tight_layout()
```
