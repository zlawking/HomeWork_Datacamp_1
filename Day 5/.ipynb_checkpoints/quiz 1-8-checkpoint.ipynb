```python
#1.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(1,10)
y=x**2
plt.plot(x,y)
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('For test')
```

```python
#2.สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู
x=np.arange(1,10)
y=np.sin(x)
plt.plot(x,y,c = 'm')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('Graph_Sin')
```

```python
#3.สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
x=np.arange(1,10)
y=x**2
plt.plot(y,x,c = '#76D7C4')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('X = Y^2')
```

```python
#4.สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -
x=np.arange(1,10)
y=np.cos(x)
plt.plot(x,y,c = 'r',ls='--')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('X = Y^2')
```

```python
#5.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น markerและเส้นสองใช้สามเหลี่ยมเป็น marker
x=np.arange(1,10)
y1=x**2
y2=x**3
plt.plot(x,y1,c = 'r',marker = 'o')
plt.plot(x,y2,c = 'b',marker = '>')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.title('2_Graph_in_1_fig')
```

```python
#6.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น markerและเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง
x=np.arange(1,10)
y=x**2
x2=y**2
plt.plot(x,y,c = 'r',marker = 'o',label='y = x^2')
plt.plot(x,x2,c = 'b',marker = '>',label='x = y^2')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.grid()
plt.legend()
plt.title('2_Graph_in_1_fig(2)')
```

```python
#7.สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกนเปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม
y=np.arange(-10,10)
x=(y**2)+(4*y)
plt.plot(x,y,c = '#FCF3CF')
plt.xlabel('axis x')
plt.ylabel('axis y')
plt.xlim([-5,120])
plt.ylim([-10,10])
plt.title('X = Y^2 + 4*Y ')
```

```python
#8.สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกันพร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend
y = np.arange(0, 10)
x = y**3
y1 = x**2 + 2
fig = plt.figure()
gr1 = fig.add_axes([0, 0, 1, 1])
gr1.plot(x, y, label= 'x = y ^3')
gr1.set_title('X = Y ^ 3')
gr1.set(xlabel = 'axis x', ylabel ='axis y')
gr1.legend(loc=0)

gr2 = fig.add_axes([0.65, 0.3, 0.35, 0.3])
gr2.plot(x, y1, label= 'y = x^2 +2')
gr2.set_title('Y = X^2 + 2')
gr2.set(xlabel = 'axis x', ylabel ='axis y')
gr2.legend(loc=0)
```
