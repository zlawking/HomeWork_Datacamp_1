```python
import numpy as np
import pandas as pd
import math
import sklearn
import sklearn.preprocessing
import datetime
import os
import matplotlib.pyplot as plt
import tensorflow as tf
import math
from math import sqrt
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
df = pd.read_csv("prices-split-adjusted.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.set_index(['date'])
```

```python
df_stock_YHOO = df[df.symbol == 'YHOO'].copy()
df_stock_YHOO.drop(['symbol'],1,inplace=True)
df_stock_YHOO.drop(['volume'],1,inplace=True)
df_stock_YHOO
```

```python
min_max_scaler = sklearn.preprocessing.MinMaxScaler()
```

```python
df_stock_YHOO['open'] = min_max_scaler.fit_transform(df_stock_YHOO.open.values.reshape(-1,1))
df_stock_YHOO['high'] = min_max_scaler.fit_transform(df_stock_YHOO.high.values.reshape(-1,1))
df_stock_YHOO['low'] = min_max_scaler.fit_transform(df_stock_YHOO.low.values.reshape(-1,1))
df_stock_YHOO['close'] = min_max_scaler.fit_transform(df_stock_YHOO.close.values.reshape(-1,1))
```

```python
seq_len = 20
valid_set_size_percentage = 10 
test_set_size_percentage = 10 
def load_data(stock, seq_len):
    data_raw = stock.to_numpy()
    data = []
    for index in range(len(data_raw) - seq_len): 
        data.append(data_raw[index: index + seq_len])
    data = np.array(data);
    valid_set_size = int(np.round(valid_set_size_percentage/100*data.shape[0]));  
    test_set_size = int(np.round(test_set_size_percentage/100*data.shape[0]));
    train_set_size = data.shape[0] - (valid_set_size + test_set_size);
    
    x_train = data[:train_set_size,:-1,:]
    y_train = data[:train_set_size,-1,:]
    
    x_valid = data[train_set_size:train_set_size+valid_set_size,:-1,:]
    y_valid = data[train_set_size:train_set_size+valid_set_size,-1,:]
    
    x_test = data[train_set_size+valid_set_size:,:-1,:]
    y_test = data[train_set_size+valid_set_size:,-1,:]
    
    return [x_train, y_train, x_valid, y_valid, x_test, y_test]

x_train, y_train, x_valid, y_valid, x_test, y_test = load_data(df_stock_YHOO, seq_len)
print('x_train.shape = ',x_train.shape)
print('y_train.shape = ', y_train.shape)
print('x_valid.shape = ',x_valid.shape)
print('y_valid.shape = ', y_valid.shape)
print('x_test.shape = ', x_test.shape)
print('y_test.shape = ',y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
#rnn
rnn_model = keras.Sequential()

rnn_model.add(layers.SimpleRNN(128, input_shape=(19, 4), return_sequences=True))
rnn_model.add(layers.SimpleRNN(64, return_sequences=False))
rnn_model.add(layers.Dense(4))
rnn_model.add(layers.Activation('tanh'))
```

```python
adam = optimizers.Adam(lr = 0.001)
rnn_model.compile(loss = 'mean_squared_error', optimizer=adam,
             metrics = ['accuracy'])
```

```python
history_rnn = rnn_model.fit(x_train, y_train, batch_size = 32,
         epochs=40, validation_data=(x_valid, y_valid))
```

```python
rnn_preds = rnn_model.predict(x_test)
col_lst = list(df_stock_YHOO.columns)
figure, axes = plt.subplots(4,1,figsize=(10, 15))

for (i, col_name) in enumerate(col_lst):
    rnn_preds_col = min_max_scaler.inverse_transform(rnn_preds[:,i].reshape(-1,1))
    Ytest = min_max_scaler.inverse_transform(y_test[:,i].reshape(-1,1))

    testScore = math.sqrt(mean_squared_error(Ytest, rnn_preds_col))
    print(col_name, ' %.2f RMSE' % (testScore))

    data = df_stock_YHOO[col_name].values.reshape(-1,1)
    data = np.reshape(data, (data.shape[0], 1))
    testPredictPlot = np.empty_like(data)
    testPredictPlot[:] = np.nan
    testPredictPlot[len(df_stock_YHOO[col_name])-len(rnn_preds_col)-1:len(df_stock_YHOO['close'])-1] = rnn_preds_col
    
    axes[i].plot(min_max_scaler.inverse_transform(df_stock_YHOO[col_name].values.reshape(-1,1)))
    axes[i].plot(testPredictPlot)
    axes[i].set_title(col_name +' Prediction')
    plt.tight_layout()
```

```python
# LSTM
lstm_model = keras.Sequential()
lstm_model.add(layers.LSTM(200, input_shape=(19, 4), return_sequences=False))

lstm_model.add(layers.Dense(4))

adam = optimizers.Adam(lr = 0.001)
lstm_model.compile(loss='mean_squared_error', optimizer=adam,
                  metrics=["accuracy"])

history_lstm = lstm_model.fit(x_train, y_train, epochs=30,
              batch_size=32, verbose=1,
              validation_data=(x_valid, y_valid))
```

```python
lstm_preds = lstm_model.predict(x_test)

col_lst = list(df_stock_YHOO.columns)
figure, axes = plt.subplots(4,1,figsize=(10, 15))

for (i, col_name) in enumerate(col_lst):
    lstm_preds_col = min_max_scaler.inverse_transform(lstm_preds[:,i].reshape(-1,1))
    Ytest = min_max_scaler.inverse_transform(y_test[:,i].reshape(-1,1))

    testScore = math.sqrt(mean_squared_error(Ytest, lstm_preds_col))
    print(col_name, ' %.2f RMSE' % (testScore))

    data = df_stock_YHOO[col_name].values.reshape(-1,1)
    data = np.reshape(data, (data.shape[0], 1))
    testPredictPlot = np.empty_like(data)
    testPredictPlot[:] = np.nan
    testPredictPlot[len(df_stock_YHOO[col_name])-len(lstm_preds_col)-1:len(df_stock_YHOO['close'])-1] = lstm_preds_col
    
    axes[i].plot(min_max_scaler.inverse_transform(df_stock_YHOO[col_name].values.reshape(-1,1)))
    axes[i].plot(testPredictPlot)
    axes[i].set_title(col_name +' Prediction')
    plt.tight_layout()
```
