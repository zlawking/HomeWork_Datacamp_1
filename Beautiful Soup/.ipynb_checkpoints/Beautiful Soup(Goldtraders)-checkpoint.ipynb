```python
import requests
import bs4
```

```python
response = requests.get('https://www.goldtraders.or.th/')
print(response.text)
```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
table = soup.find(attrs={'id':'DetailPlace_uc_goldprices1_GoldPricesUpdatePanel'})
table
```

```python
time = table.find(attrs={'id':'DetailPlace_uc_goldprices1_lblAsTime'}).string
time = str(time)
time
```

```python
text1 = table.find(attrs={'id':'DetailPlace_uc_goldprices1_lblBLSell'}).string.replace(',','')
text2 = table.find(attrs={'id':'DetailPlace_uc_goldprices1_lblBLBuy'}).string.replace(',','')
text3 = table.find(attrs={'id':'DetailPlace_uc_goldprices1_lblOMSell'}).string.replace(',','')
text4 = table.find(attrs={'id':'DetailPlace_uc_goldprices1_lblOMBuy'}).string.replace(',','')
text1 , text2 , text3 ,text4
```

```python
f = open('gold_price.csv','a')

f.write("date , GoldBar Sell, GoldBar Buy, GoldOrnament Sell, GoldOrnament Buy\n ")
f.write(time + "," + text1 + "," + text2 + "," + text3 + "," + text4 + "\n")

f.close()
```
