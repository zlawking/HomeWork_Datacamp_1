```python
import requests
import bs4
```

```python
response = requests.get('https://www.settrade.com/C13_MarketSummary.jsp?detail=SET50')
print(response.text)
```

```python
response.encoding = 'TIS-620'
```

```python
response.encoding
```

```python
soup = bs4.BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
table = soup.find(attrs={'class':'table table-info'})
table
```

```python
rows = table.find_all('td')
rows
```

```python
lst = []
for data in rows:
    lst.append(data.text)
lst
```

```python
import pandas as pd
import numpy as np
```

```python
lst_remove = []
for data in lst:
    sentence = data.replace('\r\n','').replace('\n','')
    sentence = ' '.join(sentence.split())
    lst_remove.append(sentence)
lst_remove = np.array(lst_remove)
lst_remove = lst_remove.reshape(-1,8)
lst_remove
```

```python
df = pd.DataFrame(lst_remove,columns=['','ล่าสุด','เปลี่ยนแปลง','% เปลี่ยนแปลง','สูงสุด','ต่ำสุด','ปริมาณ (\'000 หุ้น)','มูลค่า (ล้านบาท)'])
df
```
