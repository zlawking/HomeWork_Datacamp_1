4. สร้าง count plot นับข้อมูล Embarked โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
df
```

```python
sns.countplot(x='Embarked',data=df)
```

```python
sns.countplot(y='Embarked',data=df)
```

```python
sns.countplot(y='Pclass',data=df)
```

6.สร้าง box plot ให้แกน X เป็นเพศ และแกน Y เป็นราคา

7. สร้าง box plot ให้แกน X เป็นPclass และแกน Y เป็นราคา

```python
fig = plt.figure(figsize = (6,8))
sns.boxplot(x='Pclass', y = 'Fare',data=df)
```

8. สร้าง barplot โดยให้ แกน X เป็นจุดหมายปลายทางและแกน Y เป็นราคา

```python
fig = plt.figure(figsize = (6,8))
sns.barplot(x='Embarked', y = 'Fare',data=df)
```

9. สร้าง barplot โดยให้ แกน X เป็นPclass และแกน Y เป็นอายุ

```python
fig = plt.figure(figsize = (6,8))
sns.barplot(x='Pclass', y = 'Age',data=df)
```

10. สร้าง Stripplot โดยให้แกน X เป็น Pclass และ แกน Y เป็นราคา

```python
fig = plt.figure(figsize = (10,8))
sns.stripplot(x='Pclass', y = 'Fare',data=df)
```

11. สร้าง Stripplot โดยให้ แกน X เป็น Survived แกน Y เป็นอายุ และ hue เป็นเพศ

```python
fig = plt.figure(figsize = (10,8))
sns.stripplot(x='Pclass', y = 'Fare',data=df ,hue ='Sex' , dodge = True)
```

12. สร้าง Swarmplot โดยให้ แกน X เป็นเพศและแกน Y เป็นอายุ

```python
fig = plt.figure(figsize = (10,8))
sns.swarmplot(x='Sex', y = 'Age',data=df )
```

13. สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ

```python
cross_df = pd.crosstab(df['Pclass'],df['Sex'])
cross_df
```

```python
fig = plt.figure(figsize = (8,6))
sns.heatmap(cross_df , annot = True)
```

14. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ
values เป็น ราคา

```python
pivot_df = df.pivot_table(index = 'Pclass' , columns = 'Sex' , values = 'Fare')
pivot_df
```

```python
fig = plt.figure(figsize = (8,8))
sns.heatmap(pivot_df , annot = True)
```

15. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ
values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง

```python
fig = plt.figure(figsize = (8,8))
sns.heatmap(pivot_df , annot = True , linecolor = 'white' , linewidth = 2)
```

16. สร้าง Heatmap โดย Correlation ของทั้ง DataFrame

```python
Corr_df = df.corr()
Corr_df
```

```python
fig = plt.figure(figsize = (10,8))
sns.heatmap(Corr_df , cmap = 'coolwarm')
```

17. สร้าง Clustermap ของทั้ง DataFrame

```python
sns.clustermap(Corr_df , cmap = 'coolwarm')
```
