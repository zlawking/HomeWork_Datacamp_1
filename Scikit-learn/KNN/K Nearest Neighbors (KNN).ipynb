12. สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10
ratio ได้ตามใจ

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop('Class',axis=1)
y = df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

14. พิจารณาว่าควรทำ Normalization หรือ Standardization หรือไม่ควร พร้อม
ให้เหตุผล

```python
#ควรทำ Normalization เพราะ จะได้ rescale ให้ ค่า feature แต่ละตัวไม่ห่างกันเกินไป
scaler_x = MinMaxScaler()
```

```python
df_scaler_train_x = scaler_x.fit_transform(x_train)
df_scaler_test_x = scaler_x.fit_transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(df_scaler_train_x,y_train)
```

```python
predicted = knn.predict(df_scaler_test_x)
```

```python
print(confusion_matrix(y_test,predicted))
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted))
print('f1 =',f1_score(y_test,predicted,average='micro'))
print('precision =',precision_score(y_test,predicted,average='micro'))
print('recall =',recall_score(y_test,predicted,average='micro'))
```
