13.สร้าง train/test split ของข้อมูล ด้วย 70:30 ratio

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
x = df.drop(['Customer_Segment'],axis=1)
y = df['Customer_Segment']
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

14.ทำ Standardization ให้ X_train, X_test

```python
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

15.เลือกชนิดของ Baseline โมเดล 2 ชนิด (ควรเป็น Tree-based 1 อัน)

F1 score, Recall, Precision (ทั้งแบบ micro, macro)

SVM

```python
svc = SVC()
svc.fit(x_train,y_train)
```

```python
predicted_svc = svc.predict(x_test)
```

```python
print(confusion_matrix(y_test,predicted_svc))
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted_svc, 'macro'))
print('f1 =',f1_score(y_test,predicted_svc,average= 'macro'))
print('precision =',precision_score(y_test,predicted_svc,average= 'macro'))
print('recall =',recall_score(y_test,predicted_svc,average= 'macro'))
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted_svc, 'micro'))
print('f1 =',f1_score(y_test,predicted_svc,average= 'micro'))
print('precision =',precision_score(y_test,predicted_svc,average= 'micro'))
print('recall =',recall_score(y_test,predicted_svc,average= 'micro'))
```

28.อธิบายว่าการวัดผล Precision, Recall แบบ micro และ macro แตกต่างกันอย่างไร พร้อมยกตัวอย่างประกอบ

1.Micro-average Method

True positive (TP1)  = 12
False positive (FP1) = 9
False negative (FN1) = 3
Then precision (P1) and recall (R1) will be 57.14%=TP1/TP1+FP1 and 80%=TP1/TP1+FN1
and for a different set of data, the system's

True positive (TP2)  = 50
False positive (FP2) = 23
False negative (FN2) = 9
Then precision (P2) and recall (R2) will be 68.49 and 84.75


Micro-average of precision=TP1+TP2/TP1+TP2+FP1+FP2=12+50/12+50+9+23=65.96
Micro-average of recall=TP1+TP2/TP1+TP2+FN1+FN2=12+50/12+50+3+9=83.78
The Micro-average F-Score will be simply the harmonic mean of these two figures.

2.Macro-average Method

Macro-average precision=P1+P2/2=57.14+68.49/2=62.82 Macro-average recall=R1+R2/2=80+84.75/2=82.25

Suitability Macro-average method can be used when you want to know how the system performs overall across the sets of data. You should not come up with any specific decision with this average.

On the other hand, micro-average can be a useful measure when your dataset varies in size.

```python
import plotly.graph_objects as go
x_compare = ['SVM', 'RandomForest','SVM + Grid','RandomForest + Grid','SVM + PCA','RandomForest + PCA ','SVM + PCA +Grid','RandomForest + PCA + Grid']

fig = go.Figure(data=[
    go.Bar(name='Accuracy', x=x_compare, y=[accuracy_score(y_test,predicted_svc, 'macro'),accuracy_score(y_test,predicted_rf, 'macro'),accuracy_score(y_test,grid_predict_svc, 'macro'),accuracy_score(y_test,grid_predict_rf, 'macro'),accuracy_score(y_test,predict_scv_pca,'macro'),accuracy_score(y_test,predict_rf_pca, 'macro'),accuracy_score(y_test,grid_predict_pca,'macro'),accuracy_score(y_test,grid_predict_rf_pca, 'macro')]),
    go.Bar(name='F1 score', x=x_compare, y=[f1_score(y_test,predicted_svc,average= 'macro'), f1_score(y_test,predicted_rf,average= 'macro'), f1_score(y_test,grid_predict_svc,average= 'macro'), f1_score(y_test,grid_predict_rf,average= 'macro'),f1_score(y_test,predict_scv_pca,average= 'macro'),f1_score(y_test,predict_rf_pca,average= 'macro'),f1_score(y_test,grid_predict_pca,average= 'macro'),f1_score(y_test,grid_predict_rf_pca,average= 'macro')]),
    go.Bar(name='Recall', x=x_compare, y=[recall_score(y_test,predicted_svc,average= 'macro'), recall_score(y_test,predicted_rf,average= 'macro'), recall_score(y_test,grid_predict_svc,average= 'macro'), recall_score(y_test,grid_predict_rf,average= 'macro'),recall_score(y_test,predict_scv_pca,average= 'macro'),recall_score(y_test,predict_rf_pca,average= 'macro'),recall_score(y_test,grid_predict_pca,average= 'macro'),recall_score(y_test,grid_predict_rf_pca,average= 'macro')]),
    go.Bar(name='Precision', x=x_compare, y=[precision_score(y_test,predicted_svc,average= 'macro'), precision_score(y_test,predicted_rf,average= 'macro'), precision_score(y_test,grid_predict_svc,average= 'macro'), precision_score(y_test,grid_predict_rf,average= 'macro'),precision_score(y_test,predict_scv_pca,average= 'macro'),precision_score(y_test,predict_rf_pca,average= 'macro'),precision_score(y_test,grid_predict_pca,average= 'macro'),precision_score(y_test,grid_predict_rf_pca,average= 'macro')])
])

fig.update_layout(barmode='group')
fig.show()
```
