12.สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจ

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
x = df.drop('Type',axis=1)
y = df['Type']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
perdict= nb.predict(x_test)
perdict
```

```python
print(confusion_matrix(y_test,perdict))
```

```python
print('Accuracy = ',accuracy_score(y_test,perdict))
print('f1 =',f1_score(y_test,perdict,average='micro'))
print('precision =',precision_score(y_test,perdict,average='micro'))
print('recall =',recall_score(y_test,perdict,average='micro'))
```

14.พิจารณาว่าควรทำ Normalization หรือ Standardization หรือไม่ควรทั้งสองอย่าง พร้อมให้เหตุผล

15.เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

```python
#ควรทำ Normalization เพราะ จะได้ rescal ให้มี ขนาดใกล้เคียงกัน
#Normalization
scaler_x_train = MinMaxScaler()
scaler_x_test = MinMaxScaler()
```

```python
df_scaler_train_x = scaler_x_train.fit_transform(x_train)
df_scaler_test_x = scaler_x_train.fit_transform(x_test)
```

```python
nb_scaler = GaussianNB()
nb_scaler.fit(df_scaler_train_x,y_train)
```

```python
perdict_scaler = nb_scaler.predict(df_scaler_test_x)
perdict_scaler
```

```python
print(confusion_matrix(y_test,perdict_scaler))
```

```python
print('Accuracy = ',accuracy_score(y_test,perdict_scaler))
print('f1 =',f1_score(y_test,perdict_scaler,average='micro'))
print('precision =',precision_score(y_test,perdict_scaler,average='micro'))
print('recall =',recall_score(y_test,perdict_scaler,average='micro'))
```

```python
#ลองใช้ Standardization
sc_x_trian = StandardScaler()
sc_x_test = StandardScaler()
```

```python
df_scaler_Sc_train_x = sc_x_trian.fit_transform(x_train)
df_scaler_Sc_test_x = sc_x_test.fit_transform(x_test)
```

```python
nb_scaler_Sc = GaussianNB()
nb_scaler_Sc.fit(df_scaler_Sc_train_x,y_train)
```

```python
predicted_Sc = nb_scaler_Sc.predict(df_scaler_Sc_test_x)
predicted_Sc
```

```python
print(confusion_matrix(y_test,predicted_Sc))
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted_Sc,))
print('f1 =',f1_score(y_test,predicted_Sc,average='micro'))
print('precision =',precision_score(y_test,predicted_Sc,average='micro'))
print('recall =',recall_score(y_test,predicted_Sc,average='micro'))
```

16.เลือกเฉพาะ features ที่สนใจมาเทรนโมเดล และวัดผลเปรียบเทียบกับแบบ all-features

```python
x_2 = df.drop(['Type','Ca'],axis=1)
y_2 = df['Type']
```

```python
x_2_train,x_2_test,y_2_train,y_2_test = train_test_split(x_2,y_2,test_size=0.3,random_state=100)
```

```python
scaler_x_2_train = MinMaxScaler()
scaler_x_2_test = MinMaxScaler()
```

```python
df_scaler_train_x_2 = scaler_x_train.fit_transform(x_2_train)
df_scaler_test_x_2 = scaler_x_train.fit_transform(x_2_test)
```

```python
nb_2_scaler = GaussianNB()
nb_2_scaler.fit(df_scaler_train_x_2,y_2_train)
```

```python
perdict_scaler_2 = nb_2_scaler.predict(df_scaler_test_x_2)
perdict_scaler_2
```

```python
print(confusion_matrix(y_2_test,perdict_scaler_2))
```

```python
print('Accuracy = ',accuracy_score(y_2_test,perdict_scaler_2))
print('f1 =',f1_score(y_2_test,perdict_scaler_2,average='micro'))
print('precision =',precision_score(y_2_test,perdict_scaler_2,average='micro'))
print('recall =',recall_score(y_2_test,perdict_scaler_2,average='micro'))
```
