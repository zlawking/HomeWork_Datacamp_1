12.สร้าง train/test split ของข้อมูล สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจ

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
x = df.drop('Risk_good',axis=1)
y = df['Risk_good']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_train
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
predict = dtree.predict(x_test)
print(confusion_matrix(y_test,predict))
```

```python
print('Accuracy = ',accuracy_score(y_test,predict))
print('f1 =',f1_score(y_test,predict))
print('precision =',precision_score(y_test,predict))
print('recall =',recall_score(y_test,predict))
```

14.พิจารณาว่าควรทำ Normalization หรือ Standardization หรือไม่ควรทั้งสองอย่าง พร้อมให้เหตุผล

15.เลือกช้อยที่ดีที่สุดจากข้อ 14 (หรือจะทำทุกอันแล้วนำมาเปรียบเทียบก็ได้)

16.วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน Accuracy, 
F1 score, Recall, Precision

Normalization

```python
scaler_x_train = MinMaxScaler()
scaler_x_test = MinMaxScaler()
```

```python
df_scaler_train_x = scaler_x_train.fit_transform(x_train)
df_scaler_test_x = scaler_x_train.fit_transform(x_test)
```

```python
dtree_scaler = DecisionTreeClassifier()
dtree_scaler.fit(df_scaler_train_x,y_train)
```

```python
predict_scaler = dtree_scaler.predict(df_scaler_test_x)
print(confusion_matrix(y_test,predict_scaler))
```

```python
print('Accuracy = ',accuracy_score(y_test,predict_scaler))
print('f1 =',f1_score(y_test,predict_scaler))
print('precision =',precision_score(y_test,predict_scaler))
print('recall =',recall_score(y_test,predict_scaler))
```

Standardization

```python
sc_x_trian = StandardScaler()
sc_x_test = StandardScaler()
```

```python
df_scaler_Sc_train_x = sc_x_trian.fit_transform(x_train)
df_scaler_Sc_test_x = sc_x_test.fit_transform(x_test)
```

```python
dtree_scaler_Sc = DecisionTreeClassifier()
dtree_scaler_Sc.fit(df_scaler_Sc_train_x,y_train)
```

```python
predict_scaler_Sc = dtree_scaler_Sc.predict(df_scaler_Sc_test_x)
print(confusion_matrix(y_test,predict_scaler_Sc))
```

```python
print('Accuracy = ',accuracy_score(y_test,predict_scaler_Sc))
print('f1 =',f1_score(y_test,predict_scaler_Sc))
print('precision =',precision_score(y_test,predict_scaler_Sc))
print('recall =',recall_score(y_test,predict_scaler_Sc))
```

17.หาค่า parameter combination ที่ดีที่สุด สำหรับ Dataset นี้ โดยใช้ GridSearch (Hyperparameter Tuning)

```python
from sklearn.model_selection import GridSearchCV
param_combination = {'max_depth':[4,8,16,32,64],'min_samples_leaf':[1,2,4,8,16]}
```

```python
GS = GridSearchCV(DecisionTreeClassifier(),param_combination,verbose=2)
```

```python
GS.fit(df_scaler_train_x,y_train)
```

```python
GS.best_params_
```

```python
GS.best_estimator_
```

```python
grid_predict_scaler = GS.predict(df_scaler_test_x)
print(confusion_matrix(y_test,grid_predict_scaler))
```

```python
print('Accuracy = ',accuracy_score(y_test,grid_predict_scaler))
print('f1 =',f1_score(y_test,grid_predict_scaler))
print('precision =',precision_score(y_test,grid_predict_scaler))
print('recall =',recall_score(y_test,grid_predict_scaler))
```

19-21.ทำ Visualization ของค่า F1 Score,Recall,Accuracy ระหว่าง ผลลัพธ์ที่ได้จากค่า Default, ผลลัพธ์ที่ได้จากการใช้ Grid Search และ ผลลัพธ์ของ Normalization

```python
data = [[f1_score(y_test,predict),f1_score(y_test,predict_scaler_Sc),f1_score(y_test,predict_scaler),f1_score(y_test,grid_predict_scaler)],
        [recall_score(y_test,predict),recall_score(y_test,predict_scaler_Sc),recall_score(y_test,predict_scaler),recall_score(y_test,grid_predict_scaler)],
        [accuracy_score(y_test,predict),accuracy_score(y_test,predict_scaler_Sc),accuracy_score(y_test,predict_scaler),accuracy_score(y_test,grid_predict_scaler)]]
df_score = pd.DataFrame (data, columns = ['Default','Standardization','Normalization','Standardization + Grid = (The best)'],index=['F1_Score','Recall','Accuracy'])
df_score
```
