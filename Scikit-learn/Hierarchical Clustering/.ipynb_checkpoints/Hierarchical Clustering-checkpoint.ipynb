9.สร้าง Dendrogram ของ ฟีเจอร์ STG กับ PEG เพื่อหาจำนวน Cluster ที่เหมาะสม

```python
import scipy.cluster.hierarchy as sch
```

```python
x = df[['STG','PEG']].values
```

```python
plt.figure(figsize=[14,9])
dendrogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.xticks([])
```

10.สร้าง Agglomerative Hierarchical Clustering โดยใช้ฟีเจอร์ STG กับ PEG ด้วยจำนวน Cluster ที่เหมาะสม

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters= 3 ,affinity= 'euclidean',linkage='ward')
predict_hc = hc.fit_predict(x)
```

```python
predict_hc
```

11.สร้าง Clustering Visualization

```python
x1 = x[predict_hc == 0 ][:,0]
y1 = x[predict_hc == 0 ][:,1]
x2 = x[predict_hc == 1 ][:,0]
y2 = x[predict_hc == 1 ][:,1]
x3 = x[predict_hc == 2 ][:,0]
y3 = x[predict_hc == 2 ][:,1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='red',label ="Cluster 1" )
plt.scatter(x2,y2,c='blue',label ="Cluster 2" )
plt.scatter(x3,y3,c='green',label ="Cluster 3" )
plt.xlabel('STG')
plt.ylabel('PEG')
plt.legend()
```

12.เลือกฟีเจอร์ใดก็ได้ 2 ฟีเจอร์ที่ไม่ซ้ำเดิม

```python
x = df[['SCG','STR']].values
```

13.สร้าง Dendrogram ของฟีเจอร์ใหม่

```python
plt.figure(figsize=[14,9])
dendrogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.xticks([])
```

14.สร้าง Agglomerative Hierarchical Clustering ด้วย จำนวน Cluster ที่เหมาะสม โดยใช้ฟีเจอร์ใหม่

```python
hc = AgglomerativeClustering(n_clusters= 4 ,affinity= 'euclidean',linkage='ward')
predict_hc_2 = hc.fit_predict(x)
```

```python
predict_hc_2
```

15.สร้าง Clustering Visualization

```python
x1 = x[predict_hc_2 == 0 ][:,0]
y1 = x[predict_hc_2 == 0 ][:,1]
x2 = x[predict_hc_2 == 1 ][:,0]
y2 = x[predict_hc_2 == 1 ][:,1]
x3 = x[predict_hc_2 == 2 ][:,0]
y3 = x[predict_hc_2 == 2 ][:,1]
x4 = x[predict_hc_2 == 3 ][:,0]
y4 = x[predict_hc_2 == 3 ][:,1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='red',label ="Cluster 1" )
plt.scatter(x2,y2,c='blue',label ="Cluster 2" )
plt.scatter(x3,y3,c='green',label ="Cluster 3" )
plt.scatter(x4,y4,c='black',label ="Cluster 4" )
plt.xlabel('SCG')
plt.ylabel('STR')
plt.legend()
```
