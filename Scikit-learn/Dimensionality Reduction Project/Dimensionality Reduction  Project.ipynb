15.Split ข้อมูลด้วย train_test_split โดยใช้อัตราส่วน 60:40

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
x = df.drop(['Customer_Segment'],axis=1)
y = df['Customer_Segment']
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=100)
```

16.ทำ Feature Scaling ด้วย Standardization โดยใช้ข้อมูล X_train เป็นตัว fit

```python
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

17.สร้างโมเดลแบบ Support Vector Machine (เป็น Baseline)

```python
svc = SVC()
svc.fit(x_train,y_train)
```

18.สร้าง countplot ของค่าที่ทำนายโดย Support Vector Machine

```python
predicted_svc = svc.predict(x_test)
```

```python
ax = sns.countplot(x=predicted_svc)
```

19.วัดผลโมเดล SVM โดยใช้ confusion matrix

```python
print(confusion_matrix(y_test,predicted_svc))
```

20.ประเมินผลโมเดล SVM ด้วยคะแนน Accuracy, F1 score, Recall, Precision ทั้งแบบ Micro, Macro

```python
print('Accuracy = ',accuracy_score(y_test,predicted_svc, 'macro'))
print('f1 =',f1_score(y_test,predicted_svc,average= 'macro'))
print('precision =',precision_score(y_test,predicted_svc,average= 'macro'))
print('recall =',recall_score(y_test,predicted_svc,average= 'macro'))
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted_svc, 'micro'))
print('f1 =',f1_score(y_test,predicted_svc,average= 'micro'))
print('precision =',precision_score(y_test,predicted_svc,average= 'micro'))
print('recall =',recall_score(y_test,predicted_svc,average= 'micro'))
```
