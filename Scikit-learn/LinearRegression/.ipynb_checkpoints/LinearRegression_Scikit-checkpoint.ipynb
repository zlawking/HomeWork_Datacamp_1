12.สร้าง train/test split ของบ้าน สามารถลองทดสอบ 70:30, 80:20, 90:10
ratio ได้ตามใจชอบ

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x = df['Avg. Area Income']
y = df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
x_train.shape
```

13.เทรนโมเดลแบบ Simple Linear Regression

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

14.ทดสอบโมเดลวัดค่า MAE, MSE, RMSE

```python
predicted = lm.predict(x_test)
```

```python
print('MAE:',metrics.mean_absolute_error(y_test, predicted))
print('MSE:',metrics.mean_squared_error(y_test, predicted))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, predicted)))
```

15.สร้าง distribution plot เพื่อดูว่า predicted results เป็น normal ไหม

```python
plt.figure(figsize = (10,8))
sns.distplot((y_test-predicted),bins=40)
```

16.สร้าง scatter plot และ prediction line ของ simple linear regression

```python
plt.figure(figsize = (10,8))
plt.scatter(x_test,y_test,color = 'blue', label = 'data')
plt.plot(x_test,predicted ,color = 'red' , label = 'predicted Regression')
plt.legend()
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
```

17.เทรนโมเดลแบบ Multiple Linear Regression

```python
x2=df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
y2=df['Price']
```

```python
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y2,test_size=0.2,random_state=100)
```

```python
lm.fit(x2_train,y2_train)
```

```python
predicted2 = lm.predict(x2_test)
```

18.ทดสอบโมเดลวัดค่า MAE, MSE, RMSE

```python
print('MAE:',metrics.mean_absolute_error(y2_test, predicted2))
print('MSE:',metrics.mean_squared_error(y2_test, predicted2))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y2_test, predicted2)))
```

19.สร้าง distribution plot เพื่อดูว่า predicted results เป็น normal ไหม

```python
plt.figure(figsize = (10,8))
sns.distplot((y2_test-predicted2),bins=40)
```

20.ค่า RMSE ของ All-Features Multi Linear Regression มากกว่าหรือน้อย
ค่า RMSE จากคู่ที่ดีที่สุดของ Simple Linear Regression มากกว่า/น้อยกว่า
