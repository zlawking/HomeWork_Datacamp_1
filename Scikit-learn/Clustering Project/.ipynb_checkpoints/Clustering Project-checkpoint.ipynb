18.Drop Outliers ในฟีเจอร์ Apps (การทำ Clustering ควรจะจัดการ Outliers ก่อนทุกครั้ง)

```python
q1 = df['Apps'].quantile(0.25)
q3 = df['Apps'].quantile(0.75)
iqr = q3-q1
low_out  = q1-1.5*iqr
high_out = q3+1.5*iqr
```

```python
df = df.loc[(df['Apps'] > low_out) & (df['Apps'] < high_out)]
df
```

19.ทำ Min-Max Normalization ให้กับทุกฟีเจอร์ใน DataFrame

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
sc_n = MinMaxScaler()
```

```python
df_sc = sc_n.fit_transform(df)
df_sc = pd.DataFrame(df_sc,columns=df.columns)
df_sc
```

20.เลือกฟีเจอร์ Apps, Accept สร้าง 2-Means

```python
from sklearn.cluster import KMeans
```

```python
x = df_sc[['Apps','Accept']].values
```

```python
kmeans = KMeans(n_clusters=2 , random_state=99)
```

```python
kmeans_label = kmeans.fit_predict(x)
```

21.หาค่า Cluster Centers, Inertia, n_iter

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

22.สร้าง Clustering Visualization และจุด Centroids

```python
x1 = x[kmeans_label == 0 ][:,0]
y1 = x[kmeans_label == 0 ][:,1]
x2 = x[kmeans_label == 1 ][:,0]
y2 = x[kmeans_label == 1 ][:,1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='red',label ="Cluster 1" )
plt.scatter(x2,y2,c='blue',label ="Cluster 2" )
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='yellow',label ="Center" )
plt.xlabel('Apps')
plt.ylabel('Accept')
plt.legend()
```

23.สร้าง Elbow Method Visualization เพื่อหาค่า K ที่เหมาะสมที่สุด

```python
K_lst = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i , random_state=99)
    kmeans.fit(x)
    K_lst.append(kmeans.inertia_)
plt.figure(figsize = (10,8))
plt.plot(range(1,10),K_lst)
```

34.เลือกฟีเจอร์ Apps, Accept สร้าง Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
```

35.ใช้ scipy สร้าง Dendrogram ให้ method = ‘ward’

```python
plt.figure(figsize=[14,9])
dendrogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Apps, Accept')
plt.ylabel('Euclidean distances')
plt.xticks([])
```

36.สร้าง AgglomerativeClustering จากจำนวน Cluster ที่เหมาะสมที่สุดใน Dendrogram

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters= 3 ,affinity= 'euclidean',linkage='ward')
predict_hc = hc.fit_predict(x)
```

37.สร้าง Clustering Visualization ของ Hierarchical Clustering

```python
x1 = x[predict_hc == 0 ][:,0]
y1 = x[predict_hc == 0 ][:,1]
x2 = x[predict_hc == 1 ][:,0]
y2 = x[predict_hc == 1 ][:,1]
x3 = x[predict_hc == 2 ][:,0]
y3 = x[predict_hc == 2 ][:,1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x1,y1,c='red',label ="Cluster 1" )
plt.scatter(x2,y2,c='blue',label ="Cluster 2" )
plt.scatter(x3,y3,c='green',label ="Cluster 3" )
plt.xlabel('Apps')
plt.ylabel('Accept')
plt.legend()
```
